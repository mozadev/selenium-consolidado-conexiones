```python
final_combined_df = pd.read_excel(SAVE_DIR_EXTRACT_EXCEL, engine="openpyxl")
final_combined_df.head(1)
```

```python
from datetime import datetime
import os


valores_no_validos = [
                "ASIGNADO A PROYECTO CIBERSEGURIDAD",
                "DESCANSO",
                "DESCANSO MEDICO",
                "DESCANSO MÉDICO",
                "DEVOLUCIÓN DE DIA",
                "DEVOLUCIÓN DE DÍA",
                "MESA GESTIONADOS",
                "PIVOT",
                "RENUNCIA",
                "SOMBRA DE MIHAI - SOMBRA DE MIHAI - SOMBRA DE MIHAI",
                "TURNO",
                "TURNO DIA",
                "TURNO NOCHE",
                "VACACIONES",
                "VACACIONES ",
                "FERIADO",
            ]

            # Selección de columnas finales
columnas_finales = [
                'UsuarioC',
                'FechaC',
                'FechaC_formato',
                'UsuarioC_formato',
                'Hora_Inicial_General',
                'Hora_Inicial_Mesa',
                'Hora_SEMAFORO',
                'Hora_NCC',
            ]



filtrado_df = final_combined_df.copy()

columnas_hora = ['Hora_Inicial_General', 'Hora_Inicial_Mesa', 'Hora_SEMAFORO', 'Hora_NCC']
columnas_texto = ['UsuarioC', 'UsuarioC_formato']
filtrado_df = filtrado_df[~filtrado_df['Hora_Inicial_General'].isin(valores_no_validos)]
filtrado_df = filtrado_df[~filtrado_df['Hora_Inicial_Mesa'].isin(valores_no_validos)]
filtrado_df = filtrado_df[columnas_finales]
columnas_hora_existentes = [col for col in columnas_hora if col in filtrado_df.columns]
columnas_texto_existentes = [col for col in columnas_texto if col in filtrado_df.columns]
filtrado_df[columnas_hora_existentes] = filtrado_df[columnas_hora_existentes].fillna('00:00:00')
filtrado_df[columnas_texto_existentes] = filtrado_df[columnas_texto_existentes].fillna('')



filtrado_df

        
```

```python
#output_dir = 'media/reportes_combinados/'
SAVE_DIR_EXTRACT_EXCEL = BASE_DIR / "media" / "reportes_combinados" / 'filtrado'
os.makedirs(SAVE_DIR_EXTRACT_EXCEL, exist_ok=True) 
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
ruta = os.path.join(SAVE_DIR_EXTRACT_EXCEL, f'df_filtrado_anexos{timestamp}.xlsx')
filtrado_df.to_excel(ruta, index=False, engine='openpyxl')
print(f"Reporte Combinado guardado en: {ruta}")
```
